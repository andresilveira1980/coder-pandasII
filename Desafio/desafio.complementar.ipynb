```python
consoles_por_companhia = df.groupby('Company').size()
consoles_por_companhia
```

```python
market_values = df.groupby('Company')['Market Value'].first()
market_values
```

```python
consoles_market_df = pd.DataFrame({
    'Consoles Count': consoles_por_companhia,
    'Market Value': market_values
})
```

```python
consoles_market_df.head
```

```python
consoles_market_df_sorted = consoles_market_df.sort_values(by='Consoles Count', ascending=False)
```

```python
consoles_market_df_sorted
```

```python
consoles_market_df_sorted['Average Market Value per Console'] = consoles_market_df_sorted['Market Value'] / consoles_market_df_sorted['Consoles Count']
```
